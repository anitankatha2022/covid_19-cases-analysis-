Removing the missing values

```python
data=dataset.dropna()
```

Creating variablesf for ease acccess

```python
Western_Pacific = data[data['Region'] == 'Western Pacific']
East_Medi = data[data['Region'] == 'Eastern Mediterranean']
Africa = data[data['Region'] == 'Africa']
Americas = data[data['Region'] == 'Americas']
SE_Asia = data[data['Region'] == 'South-East Asia']
Europe = data[data['Region'] == 'Europe']
```

Data visualisation

```python
plt.figure(figsize=(7,8))
plt.ticklabel_format(useOffset=False, style='plain')
sns.barplot(x=data['Region'], y=data['Total Deaths'], errwidth=0)
plt.xticks(rotation=75, size=14)
plt.yticks(size=14)
plt.xlabel('Regions', fontsize=20)
plt.ylabel('Total Deaths', fontsize=20)
plt.title(' Total Deaths according to the Regions', fontsize=20, x=0.4, y=1.04)
plt.show()
```

```python
plt.figure(figsize=(7,9))
plt.ticklabel_format(useOffset=False, style='plain')
sns.barplot(x=data['Region'], y=data['Total Cases'], errwidth=0)
plt.xticks(rotation=75, size=14)
plt.yticks(size=14)
plt.xlabel('Regions', fontsize=20)
plt.ylabel('Total Cases', fontsize=20)
plt.title(' Total cases based on the Regions', fontsize=20, x=0.4, y=1.04)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.ticklabel_format(useOffset=False, style='plain')
sns.barplot(x=Africa['Country'], y=data['Total Deaths'])
plt.xticks(rotation=90, size=18)
plt.yticks(size=18)
plt.xlabel('Country', fontsize=30)
plt.ylabel('Total Deaths', fontsize=30)
plt.title('Total Deaths based on the Africa Region', fontsize=30, x=0.3, y=1.04)
plt.show()
```

```python
plt.figure(figsize=(30,20))
plt.ticklabel_format(useOffset=False, style='plain')
sns.barplot(x=SE_Asia['Country'], y=data['Total Cases'])
plt.xticks(rotation=90, size=18)
plt.yticks(size=18)
plt.xlabel('Country', fontsize=30)
plt.ylabel('Total Cases', fontsize=30)
plt.title(' Total Cases based on the South East Asia Region', fontsize=30, x=0.3, y=1.04)
plt.show()
```
